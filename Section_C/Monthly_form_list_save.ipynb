```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
df = pd.read_csv('yds_train2018.csv')
df.head()
```

```python
df
```

```python
mylist = []
```

```python
mylist2 = []
```

```python
##DATA CLEANING
## This was done in a loop to get total monthly sales(considering weeks and merchants both..summing them up)
for jx in range(2015,2017):
    for ix in range(1,13):
        a = df.loc[(df['Year']==jx) & (df['Month']==ix) & (df['Country']=='Columbia') & (df['Product_ID']==3)]
        #print(a)
        y = a.iloc[0]['S_No']
        x = a['Sales'].sum(axis=0)
        a.loc[y-1,'Sales']=x
        a.drop(['S_No','Merchant_ID','Week'],axis=1,inplace=True)
        mylist2.append(np.asarray(a.loc[y-1]))
np.asarray(mylist2)
```

```python
import pickle

with open('outfile', 'wb') as fp:
    pickle.dump(mylist2, fp)
```
