```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
data = pd.read_csv('final_train_data.csv')
```

```python
data
```

```python
countries = np.unique(data['Country'])
```

```python
country_map = {
                'Argentina':0,
                'Belgium':1,
                'Columbia':2,
                'Denmark':3,
                'England':4,
                'Finland':5 }
```

```python
data.Country = data.Country.map(country_map)
data = shuffle(data)
data
```

```python
X = data.drop('Sales',axis = 1)
Y = data['Sales']
X.shape, Y.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X, Y)
```

```python
lm.score(X,Y)
```

```python
test_df = pd.read_csv('yds_test2018.csv')
test_df.drop('S_No',axis=1,inplace=True)
```

```python
test_df
```

```python
test_df.Country = test_df.Country.map(country_map)
test_df
```

```python
X_test = test_df.drop('Sales',axis=1)
X_test.shape
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
test_df['Sales']=predictions
test_df
```

```python
test_df.to_csv('Linear_Regression_predictions.csv')
```
